```python
# ðŸ“¥ Step 1: Load live NYPD Arrest Data
import pandas as pd

url = "https://data.cityofnewyork.us/api/views/uip8-fykc/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

# ðŸ§¹ Step 2: Clean AGE_GROUP and filter valid entries
valid_ages = ['<18', '18-24', '25-44', '45-64', '65+']
df = df[df['AGE_GROUP'].isin(valid_ages)]

# ðŸ—ºï¸ Step 3: Filter boroughs - Manhattan (M) & Bronx (B)
df_m = df[df['ARREST_BORO'] == 'M']
df_b = df[df['ARREST_BORO'] == 'B']

# ðŸ“Š Step 4: Arrest counts by age group
m_counts = df_m['AGE_GROUP'].value_counts().reindex(valid_ages, fill_value=0)
b_counts = df_b['AGE_GROUP'].value_counts().reindex(valid_ages, fill_value=0)
# ðŸ“‹ Step 5: Descriptive statistics
print("Descriptive Statistics for Arrest Age Groups:\n")

desc_stats = pd.DataFrame({
    'Age Group': valid_ages,
    'Manhattan Arrests': m_counts.values,
    'Bronx Arrests': b_counts.values
})

desc_stats['Total'] = desc_stats['Manhattan Arrests'] + desc_stats['Bronx Arrests']
desc_stats['Percent Bronx'] = round((desc_stats['Bronx Arrests'] / desc_stats['Total']) * 100, 1)

print(desc_stats)
# ðŸ“ˆ Step 6a: Bar plot - Manhattan vs Bronx
import seaborn as sns
import matplotlib.pyplot as plt

plot_df = desc_stats.melt(id_vars='Age Group',
                          value_vars=['Manhattan Arrests', 'Bronx Arrests'],
                          var_name='Borough', value_name='Count')

plt.figure(figsize=(10, 6))
sns.barplot(data=plot_df, x='Age Group', y='Count', hue='Borough')
plt.title('Arrests by Age Group: Manhattan vs. Bronx')
plt.xlabel('Age Group')
plt.ylabel('Number of Arrests')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
# ðŸ¥§ Step 6b: Pie chart - Age group distribution in the Bronx
plt.figure(figsize=(7, 7))
plt.pie(b_counts, labels=valid_ages, autopct='%1.1f%%', startangle=140)
plt.title('Age Group Proportion of Arrests in the Bronx')
plt.axis('equal')
plt.show()
# ðŸ§  Step 7: Conclusion
print("""
Conclusion:
Descriptive stats show the 25â€“44 age group leads in arrest counts across both boroughs.
However, the Bronx shows a relatively higher share of younger arrestees (<25) compared to Manhattan.
Visualizations highlight both volume and proportion differences across boroughs.
Bias may arise not from actual crime distribution, but from concentrated policing in certain communities.
""")

```
